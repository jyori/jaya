```python
url = 'http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population'
store_html = requests.get(url).text
#print store_html
```

```python
def population_data(html):
    dom = web.Element(html)
    tables_in = dom.by_class('sortable wikitable')
    return tables_in

population_tables = population_data(store_html)
print "table length: %d" %len(population_tables)
for t in population_tables:
    print t.attributes
```

```python

def all_population(population_tables):
    
    result = defaultdict(dict)
    
    for tbl in population_tables:
        table_header = tbl('tr') #get all table rows
        get_first_row = table_header[0] # navigate to first row
        get_header = get_first_row('th') # get the headers

        year_value = [int(i.content) for i in get_header if i.content.isnumeric()] # isnumeric returns only the numeric content.
#        print year_value
        year_indices = [idx for idx,i in enumerate(get_header) if i.content.isnumeric()] 

        rows = tbl('tr')[1:]
        for row in rows:
            get_td = row('td')
            country_name = get_td[0]('a')[0].content
            population = [int(get_td[i].content.replace(',', '')) for i in year_indices]
 #           print population
            another_dict = dict(zip(year_value, population))
            result[country_name].update(another_dict)
    return result
            
result = all_population(new_dict[u'Country or territory'])
print len(result), "Countries Extracted"
print "Amazing Afhgan"
print result[u'Afghanistan']
```
